```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
```

```python

california_housing = fetch_california_housing(as_frame=True).frame

correlation = california_housing.corr()
print(correlation)

```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# prepare the data
X = california_housing['MedInc'].values.reshape(-1, 1)
y = california_housing['MedHouseVal'].values

# fit the linear regression model
model = LinearRegression()
model.fit(X, y)

import matplotlib.pyplot as plt
```

```python

# plot the data
plt.scatter(X, y)

# plot the best fit line
plt.plot(X, model.predict(X), color='red')

# set the axis labels and title
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.title('California Housing')

# show the plot
plt.show()
```

```python

# compute the predicted values
y_pred = model.predict(X)

mse = np.mean((y - y_pred)**2)
rmse = np.sqrt(mse)
print("Root Mean Squared Error:", rmse)
```
